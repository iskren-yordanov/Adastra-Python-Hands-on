define a BlackJackCard class and a Deck class so that the following code (which draws 3 cards) will work:

print(h.total)

```python
from enum import Enum
from typing import List
import random

class Suit(str,Enum):
    Club="♣"
    Diamond="♦"
    Heart="♥"
    Spade="♠"


class BlackJackCard:
    def __init__(self, name, suit, soft, hard):
        self.name = name
        self.suit = suit
        self.soft = soft
        self.hard = hard

    def __str__(self)->str:
        return f"({self.name} {self.suit} {self.hard} {self.soft})"

class Deck():

    def __init__(self):

        self.deck_cards = []

        card_numbers = [2,3,4,5,6,7,8,9,10, 'King', 'Queen', 'Jack', 'Ace']
        # should generate and randomize the set:
        for n in card_numbers:
            for s in Suit:

                if str(n) in ("King", "Queen", "Jack"):
                    hard = soft = 10
                elif str(n) in ("Ace"):
                    hard = 1
                    soft = 11
                else:
                    hard = soft = n
                crd = BlackJackCard(str(n), s.value, soft, hard)
                self.deck_cards.append(crd)
        
        random.shuffle(self.deck_cards)
            

    def __str__(self)->str:
        return " | ".join(map(str,self.deck_cards)) # self.card referes to the Hand_Lazy implementation

    def pop(self) -> BlackJackCard:
        return self.deck_cards.pop()

class Hand:
    def __init__(
            self,
            dealer_card:BlackJackCard,
            *cards:BlackJackCard
        ) -> None:
        self.dealer_card = dealer_card
        self._cards = list(cards)
    
    def __str__(self)->str:
        return",".join(map(str,self.card)) # self.card referes to the Hand_Lazy implementation
    
    def __repr__(self)->str:
        return(
            f"{self.__class__.__name__}"
            f"({self.dealer_card!r},"
            f"{','.join(map(repr,self.card))})"
        )

class Hand_Lazy(Hand):
    
    @property
    def total(self)->int:
        delta_soft=max(c.soft - c.hard for c in self._cards)
        hard_total=sum(c.hard for c in self._cards)
        
        if hard_total + delta_soft<=21:
            return hard_total + delta_soft
        
        return hard_total

    @property
    def card(self) -> List[BlackJackCard]:
        return self._cards
    
    @card.setter
    def card(self,aCard:BlackJackCard) -> None:
        self._cards.append(aCard)

    @card.deleter
    def card(self) -> None:
        self._cards.pop(-1)



#tmp = BlackJackCard(Suit.Spade.value, 2,2)
#print(tmp)

d=Deck()
#print(d)

h = Hand_Lazy(d.pop(),d.pop(),d.pop())
print(h.total)


for c in h.card:
    print(f"we have: {c}")

print(f"current deader card: {h.dealer_card}")


del h.card

h.card = d.pop()

for c in h.card:
    print(f"we have: {c}")

print(f"current deader card: {h.dealer_card}")
```

```python
# creating a class
class Person:
    
    def __init__(self, name):
        self._name = name

    # using decorator for property
    @property
    def name(self):
        return self._name

    # using decorators for setter
    @name.setter
    def name(self, value):
        self._name = value

    #using decorator for python deleter
    @name.deleter
    def name(self):
        del self._name


# creating new object

person2 = Person('Ivan')

person = Person('Peter')
person.name = "Iskren"

# deleting the name 
#del person.name

# printing the name
print(person.name, person2.name)

#The deleter function in Python is used to remove a class’s properties. The property cannot be accessed using the same instance after being deleted.
```
